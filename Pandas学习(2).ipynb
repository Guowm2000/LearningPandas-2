## **6. Pandas缺失值处理**

Pandas使用以下函数处理缺失值：  

- **isnull 和 notnull**: 检测是否为空值，可用于 df 和 Series
- **dropna**: 丢弃删除缺失值
    - axis: 删除行还是列，{0 or 'index',1 or 'columns'},default 0
    - how: 如果等于any，则行或列中包含任意空值，则该行或列都会被删除，如果等于all，则该行或列的所有值为空时才删除该行或列
    - inplace: 如果为True则修改当前df，否则返回新的df
- **fillna**: 填充空值，即将空值填充为指定的值,在一行或一列中遇到空值时该函数才执行。
    - value: 用于填充的值，可以是单个值，或是字典{key是列名，value是这一列将要填充的值}
    - method: 等于 ffill (forward fill),则用这一列或这一行前一个不为空的值填充，等于 bfill (backward fill),则使用后一个不为空的值填充
    - axis: 按行还是按列填充{0 or 'index',1 or 'columns'}
    - inplace: 如果为True则修改当前df，否则返回新的df

## **7. Pandas的SettingWithCopyWarning报警**

### **7.0 读取数据**

```python
import pandas as pd
```

```python
df=pd.read_csv('gy202302.csv')
```

### **7.1 复现该报警**

```python
#选出企业名称（H02）以‘上海’开头的数据
condition=df['H02'].str.startswith('上海')
```

```python
condition.value_counts()
```

```python
df[condition]['zf']=df[condition]['N2000_2']/df[condition]['N2000_4']-1
```

**上面出现了SettingWithCopyWarning报警**

### **7.2 原因分析**

df[condition]['zf']是通过两部操作完成的，相当于  

`df.get(condition).set(df[condition]['N2000_2']/df[condition]['N2000_4']-1)`  

报警是在get时发出的

由于get得到的dataframe可能是view，也可能是 copy，所以pandas发出了报警

**核心要诀： 针对DataFrame的写操作，只允许在源dataframe上操作，且一步完成**

### **7.3 解决方法**

方法1： 将get加set的两步操作，改为set的一步操作

```python
df.loc[condition,'zzff']=df[condition]['N2000_2']/df[condition]['N2000_4']-1
```

方法2： 将第一步得到的结果存储到新的dataframe中，然后在对它进行set

```python
df1=df[condition].copy()
```

```python
df1['zzff']=df[condition]['N2000_2']/df[condition]['N2000_4']-1
```

Series的排序：  
**Series.sort_values(ascending=True,inplace=False)**  
参数说明：  
- ascending： 默认为True，表示升序，False表示降序  
- inplace: 是否修改原始Series  

DataFrame的排序：  
**DataFrame.sort_values(by,ascending=True,inplace=False)**  
参数说明：  
- by: 字符串，或者List<字符串>，单列排序或者多列排序
- ascending: bool或者List(此list中是bool值),升序还是降序，如果是List，则对应by的多列，分别是升序还是降序
- inplace: 是否修改原始DataFrame

## **10. Pandas的axis参数怎么理解**  

- axis=0或'index':
    - 如果是单行操作，指的就是一行
    - 如果是聚合操作，指的就是跨行，cross rows
- axis=1或'columns':
    - 如果是单列操作，指的就是某一列
    - 如果是聚合擦作，指的就是跨列，cross columns

**总结：指定按哪个axis，就是这个axis要动起来（类似被for循环遍历），其他的axis保持不动**

```python
import pandas as pd
import numpy as np
```

```python
#自定义一个dataframe
df=pd.DataFrame(
    np.arange(12).reshape(3,4),   #指定数据
    columns=['A','B','C','D']     #指定列名
)
                
```

```python
df
```

```python
df.drop('A',axis=1)
```

### **10.2 单行drop，就是删除某一行**

```python
df.drop(1,axis=0)
```

### **10.3 按axis=0/index执行mean聚合操作**  

反直觉：输出的不是每行的结果，而是每列的结果

```python
df
```

```python
df.mean(axis=0)  #axis=0表示对各行进行for循环
```

**指定按哪个axis，就是这个axis要动起来（类似被for循环遍历），其他的axis保持不动**

### **10.5 再次举例，加深理解**  


**指定按哪个axis，就是这个axis要动起来（类似被for循环遍历），其他的axis保持不动**

```python
def get_sum_value(x):
    return x['A']+x['B']+x['C']+x['D'] # 由于定义了axis=1,所以x是一个横向（跨列）的series

df['sum_value']=df.apply(get_sum_value,axis=1) ##此处axis=1表示传给函数的参数x会跨列，即可用x['A']取到列值
```
